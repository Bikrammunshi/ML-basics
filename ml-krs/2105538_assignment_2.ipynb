```python
#q12
df['blood_donation']=df['whether he/she donated blood in March 2007']
df.drop('whether he/she donated blood in March 2007',axis=1,inplace=True)
df
```

```python
#q13
Df=df.to_numpy()
plt.hist(Df)
```

```python
#q14

plt.bar(np.arange(0,748),df['blood_donation']>0,color='r',alpha=1,label='Donated')
plt.legend()
```

```python
#q15
plt.figure(figsize=(10,10))
ax = plt.axes(projection ="3d")
ax.scatter3D(df['Recency (months)'], df['Frequency (times)'], df['blood_donation'], color = "#40E0D0")
plt.title("KRS")
ax.set_xlabel('Recency (months)')
ax.set_ylabel('Frequency (times)')
ax.set_zlabel('blood_donation')
plt.legend()
```

```python
#q16
plt.figure(figsize=(20,20))
plt.pie(df.mean(axis=0).to_numpy(),labels=['Recency (months)',	'Frequency (times)'	,'Monetary (c.c. blood)'	,'Time (months)'	,'blood_donation'],colors=['#40E0D0','g','b','y','r'])
plt.legend()
```

```python
#q17
plt.figure(figsize=(10,10))
Df=df.drop(['blood_donation'],axis=1).to_numpy()
plt.boxplot(Df,labels=['Recency (months)','Frequency (times)','Monetary (c.c. blood)','Time (months)'])
```

```python
#q18
plt.figure(figsize=(10,10))

plt.bar(df['Time (months)'], df['blood_donation'], color ='#000080',width =1)
plt.xlabel('Time (months)')
plt.ylabel('blood_donation')
plt.title('Time Vs Blood Donation')
```
