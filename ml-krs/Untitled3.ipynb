```python
    take=Sales[['Profit','Region','Product_Category','Order_Priority']]
output=Sales['Sales']
```

```python
model.fit(x_train,y_train)
```

```python
predict_train=model.predict(x_train)
predict_test=model.predict(x_test)
predict_train
```

```python
from sklearn import metrics
```

```python
score=metrics.r2_score(y_test,predict_test)
scores=[]
for i in range(10000):
    x_train,x_test,y_train,y_test=train_test_split(take,output,test_size=0.2,random_state=i)
    model.fit(x_train,y_train)
    predict_test=model.predict(x_test)
    score=metrics.r2_score(y_test,predict_test)
    scores.append(score)
print(max(scores))
print(scores.index(max(scores)))
```
