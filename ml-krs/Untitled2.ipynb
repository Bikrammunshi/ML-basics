```python
given=fish.iloc[:,2:7]
output=fish.iloc[:,1]
```

```python
given
```

```python
output
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(given,output,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict_train=model.predict(x_train)
predict_test=model.predict(x_test)
predict_train
```

```python
from sklearn import metrics
```

```python
score=metrics.r2_score(y_test,predict_test)
scores=[]
for i in range(5000):
    x_train,x_test,y_train,y_test=train_test_split(given,output,test_size=0.2,random_state=i)
    model.fit(x_train,y_train)
    predict_test=model.predict(x_test)
    score=metrics.r2_score(y_test,predict_test)
    scores.append(score)
print(max(scores))
print(scores.index(max(scores)))
```
