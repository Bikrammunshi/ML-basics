```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('linear_regression.csv')
df
```

```python
#encoding

df.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
df.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
df.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
df
```

```python
#splitting a panda dataframe
x=df.drop(['Car_Name','Selling_Price'],axis=1)
y=df['Selling_Price']
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
x_test
```

```python
x_train
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics #for accuracy and error calculation
```

```python
#Load the model
lr=LinearRegression()
```

```python
lr
```

```python
lr.fit(x_train,y_train)
```

```python
train_pred=lr.predict(x_train)
test_pred=lr.predict(x_test)
```

```python
train_pred
```

```python
test_pred
```

```python
#r squared error

error_train=metrics.r2_score(y_train,train_pred)
error_train
```

```python

error_test=metrics.r2_score(y_test,test_pred)
error_test
```

```python
plt.figure(figsize=(5,5))
plt.scatter(train_pred,y_train,color='brown')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(test_pred,y_test,color='purple')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```
